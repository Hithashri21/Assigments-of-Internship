```python
# Assignment 03 - Build Your First Dataset

# Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Create Dataset (Study Hours vs Marks)
# Feature (X) → Study Hours
# Label (y) → Marks

X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([30, 40, 50, 60, 70, 75, 85, 90])

# Step 3: Create Linear Regression Model
model = LinearRegression()

# Step 4: Train the Model
model.fit(X, y)

# Step 5: Predict Marks
predicted_marks = model.predict(X)

# Step 6: Predict for new values (9 and 10 hours)
new_hours = np.array([9, 10]).reshape(-1, 1)
new_predictions = model.predict(new_hours)

print("Predicted Marks for 9 and 10 hours:", new_predictions)

# Step 7: Plot the Dataset and Regression Line
plt.scatter(X, y, label="Actual Data")
plt.plot(X, predicted_marks, label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.legend()
plt.show()
```
