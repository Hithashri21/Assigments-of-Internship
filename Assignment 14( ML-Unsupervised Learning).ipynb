```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate dataset (acts like mall customers)
X, _ = make_blobs(n_samples=200, centers=5, cluster_std=1.8, random_state=42)

# Convert to DataFrame
df = pd.DataFrame(X, columns=['Annual Income', 'Spending Score'])

print(df.head())

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Cluster Centers
centers = kmeans.cluster_centers_

# Plot Clusters
plt.scatter(df['Annual Income'], df['Spending Score'], c=df['Cluster'])
plt.scatter(centers[:,0], centers[:,1], s=200, marker='X')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()
```
