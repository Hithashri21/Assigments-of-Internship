```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances

# User preferences (ratings)
# [Movie A, Movie B]
aarti = np.array([[5, 4]])
user1 = np.array([[5, 4]])
user2 = np.array([[1, 2]])

# Calculate distances
dist1 = euclidean_distances(aarti, user1)
dist2 = euclidean_distances(aarti, user2)

print("Distance with User1:", dist1)
print("Distance with User2:", dist2)

# Recommendation logic
if dist1 < dist2:
    print("Recommend movies liked by User1")
else:
    print("Recommend movies liked by User2")
```
