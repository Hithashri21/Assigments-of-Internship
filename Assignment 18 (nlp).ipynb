```python
import re

# Sample messy sentences
sentences = [
    "OMG this movie is lit 🔥🔥",
    "I cant beleive this happend",
    "LOL that was sooo funny 😂",
    "u r amazing bro",
    "This is gr8 news!!!",
    "What r u doing???",
    "I luv this song ❤️",
    "Plz send me the details ASAP",
    "He dont know anything",
    "Im sooo tired rn",
    "This product is awsm",
    "Where r u???",
    "Gud morning everyone",
    "I didnt like the food 😒",
    "Btw I'll call u later",
    "Thx for ur help 🙏",
    "This is not gud at all",
    "Idk what to do now",
    "She is sooo pretty!!!",
    "Wat is this nonsense"
]

# Slang dictionary
slang_dict = {
    "u": "you",
    "r": "are",
    "lol": "",
    "omg": "",
    "gr8": "great",
    "luv": "love",
    "plz": "please",
    "asap": "as soon as possible",
    "btw": "by the way",
    "thx": "thanks",
    "ur": "your",
    "idk": "i do not know",
    "rn": "right now",
    "awsm": "awesome",
    "gud": "good",
    "wat": "what"
}

# Function to clean text
def clean_text(text):
    text = text.lower()  # convert to lowercase

    # remove emojis and punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # replace slang words
    words = text.split()
    words = [slang_dict.get(word, word) for word in words]
    text = " ".join(words)

    # normalize repeated letters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    return text

# Apply preprocessing
for sentence in sentences:
    print("Original :", sentence)
    print("Cleaned  :", clean_text(sentence))
    print("-" * 50)
```
