```python
def assign_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    elif marks >= 50:
        return "D"
    else:
        return "F"
students = [
    {"name": "Rahul", "marks": 85},
    {"name": "Anjali", "marks": 92},
    {"name": "Kiran", "marks": 76},
    {"name": "Sneha", "marks": 64},
    {"name": "Arjun", "marks": 88}
]
for student in students:
    student["grade"] = assign_grade(student["marks"])
total_marks = sum(student["marks"] for student in students)
average = total_marks / len(students)
topper = max(students, key=lambda x: x["marks"])
print("Student Details:")
for student in students:
    print(f"Name: {student['name']}, Marks: {student['marks']}, Grade: {student['grade']}")

print("\nClass Average:", round(average, 2))
print("Topper:", topper["name"], "with", topper["marks"], "marks")
```
