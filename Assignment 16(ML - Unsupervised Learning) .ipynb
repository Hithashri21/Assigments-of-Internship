```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic mall-like data
X, _ = make_blobs(n_samples=200, centers=5, cluster_std=1.5, random_state=42)

df = pd.DataFrame(X, columns=['Income', 'Spending'])

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Plot
plt.scatter(df['Income'], df['Spending'], c=df['Cluster'])
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=200, marker='X')
plt.title("Customer Segmentation")
plt.xlabel("Income")
plt.ylabel("Spending")
plt.show()
```
